# Importing G2 Library

```python
from G2lib import G2ProductFeatureList

from dotenv import load_dotenv
import os

load_dotenv()
token = os.getenv("API_KEY")
```

# Default Feature List

```python
g2_default  = G2ProductFeatureList(token)
g2_default.run()
```

```python
len(g2_default.review_dict) 
```

# Time Bounded Feature List

```python
g2_time = G2ProductFeatureList(token)
g2_time.time_features_extraction("2021-01-01T00:00:00Z")
```

```python
len(g2_time.review_dict)
```

```python
g2_country = G2ProductFeatureList(token)
g2_country.geographic_features("United States")
```

```python
len(g2_country.review_dict)
```
